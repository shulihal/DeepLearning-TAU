# Q4
### (a) Prepare the data to check the solution

```python
n=10000
rng = np.random.default_rng(seed=42)
X = rng.random(size=(n, 4))
e = rng.standard_normal(size=(n))

Y=X[:,0]-2*X[:,1]+3*X[:,2]-4*X[:,3]+e
```

### (c) Implement the Gradient Descent (GD) algorithm

```python
ind = rng.integers(0,n,n)
X[ind]
Y[ind]
```

```python
W = np.zeros(5) #initial guess
X_ = np.c_[np.ones(n), X]  #adding ones for the bias

def gradient(w, x, y): #gradient of RSS for given weights
    return (2*x.T*(x@w-y)).sum(axis=1)/n
```

```python
import matplotlib.pyplot as plt
plt.plot(mse)
```

```python
alpha=0.01  #step size

mse = np.array([])

for i in range(10000):
    ind = rng.integers(0,n,n)
    Xsam=X_[ind]
    Ysam = Y[ind]
    
    RSS = (Xsam@W-Ysam).T@(Xsam@W-Ysam)
    np.append(mse, RSS/n)
    if i%100==0:
        print(RSS/n)
    W -= alpha*gradient(W, Xsam, Ysam) # a step
```

```python
for i in range(100000):
    RSS = (X_@W-Y).T@(X_@W-Y)
    if i%10000==0:
        print(RSS/n)
    W -= alpha*gradient(W) # a step
    alpha= alpha*math.exp(-alpha) #exponential decay
```

```python
gamma = 0.8
v=0
alpha=0.1
for i in range(10000):
    RSS = (X_@W-Y).T@(X_@W-Y)
    if i%1000==0:
        print(RSS/n)
    v= v*gamma + alpha*gradient(W)
    W -= v
```
